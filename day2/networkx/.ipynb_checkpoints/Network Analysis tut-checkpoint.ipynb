## Do you know Each other

```python
H = nx.complete_graph(5)
```

```python
nx.draw(H, with_labels=True)
```

```python
nx.triangles(H)
```

```python
import itertools
```

```python
def know_this_person():
    open_triangle = nx.path_graph(3)
    combi = list(itertools.combinations(H.nodes(), 3))
    result = []
    for i in  combi:
        if nx.is_isomorphic(nx.subgraph(H, i), open_triangle):
            result.append(i)
        print(result)
```

```python
know_this_person()
```
